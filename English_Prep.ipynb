## 1. The Sentence: Building Blocks of Language

A sentence is a group of words that expresses a complete thought. It must contain a **subject** and a **predicate**.

### Subject and Predicate

* **Subject**: Who or what the sentence is about. It's usually a noun or a pronoun.
* **Predicate**: What the subject is or does. It contains the verb.

**Examples**:
1.  **Subject**: `The judge`
    **Predicate**: `listened to the argument.`
2.  **Subject**: `All the legal documents`
    **Predicate**: `are on the table.`
3.  **Subject**: `She`
    **Predicate**: `is studying contract law.`

### Phrase and Clause

* **Phrase**: A group of words that acts as a single part of speech but **does not** have a subject and a verb.
    * **Examples**: `in the courtroom`, `with a valid point`, `the senior advocate`.

* **Clause**: A group of words that **does have** a subject and a verb.
    * **Independent Clause**: Expresses a complete thought and can stand alone as a sentence.
        * **Example**: `The evidence was presented.`
    * **Dependent (or Subordinate) Clause**: Does not express a complete thought and cannot stand alone. It often starts with words like `because`, `although`, `when`, `which`, `that`.
        * **Example**: `...because the witness was absent.`

**Putting it all together**:
* `The trial was postponed` (Independent Clause) `because the witness was absent` (Dependent Clause).

## 2. Noun Properties: Case, Number, Person, Gender

### Case
Case shows the function of a noun or pronoun in a sentence.
* **Subjective (Nominative) Case**: The noun/pronoun is the subject of the verb.
    * **Examples**: `He` filed the lawsuit. `The lawyer` presented the facts.
* **Objective (Accusative) Case**: The noun/pronoun is the direct object, indirect object, or object of a preposition.
    * **Examples**: The judge overruled `him`. She gave the file to `the clerk`.
* **Possessive (Genitive) Case**: The noun/pronoun shows ownership.
    * **Examples**: This is the `client's` file. The decision is `hers`.

### Number
* **Singular**: Refers to one person, place, or thing. (e.g., `witness`, `case`, `child`)
* **Plural**: Refers to more than one. (e.g., `witnesses`, `cases`, `children`)

### Person
* **First Person**: The speaker. (`I`, `we`)
* **Second Person**: The person being spoken to. (`you`)
* **Third Person**: The person or thing being spoken about. (`he`, `she`, `it`, `they`, `the advocate`)

### Gender
* **Masculine**: `man`, `king`, `he`
* **Feminine**: `woman`, `queen`, `she`
* **Common**: Can refer to either male or female. (`lawyer`, `judge`, `student`, `they`)
* **Neuter**: Refers to things without gender. (`book`, `courtroom`, `it`)

## 3. Tense: Locating Actions in Time

Tense indicates the time of an action or event.

| Tense | Form | Example |
| :--- | :--- | :--- |
| **Present Simple** | `Verb (s/es)` | He **argues** his case well. The court **opens** at 10 AM. |
| **Present Continuous**| `is/am/are + V-ing` | The lawyer **is presenting** the evidence now. |
| **Present Perfect** | `has/have + V3` | She **has filed** three petitions this week. |
| **Present Perfect Continuous** | `has/have + been + V-ing`| They **have been debating** this point for an hour. |
| **Past Simple** | `Verb (2nd form)` | The judge **delivered** the verdict yesterday. |
| **Past Continuous** | `was/were + V-ing` | He **was reading** the case file when the phone rang. |
| **Past Perfect** | `had + V3` | The plaintiff **had** already **left** before the lawyer arrived. |
| **Past Perfect Continuous** | `had + been + V-ing` | They **had been negotiating** for months before they reached a settlement. |
| **Future Simple** | `will/shall + V1` | The court **will hear** the case tomorrow. |
| **Future Continuous**| `will/shall + be + V-ing` | We **will be waiting** for the judgment at noon. |
| **Future Perfect** | `will/shall + have + V3`| By next year, she **will have completed** her law degree. |
| **Future Perfect Continuous** | `will/shall + have + been + V-ing`| By 2030, he **will have been practicing** law for 20 years. |

## 4. Active and Passive Voice

* **Active Voice**: The subject **performs** the action. (Structure: **Subject + Verb + Object**)
    * **Example**: `The police arrested the suspect.`
* **Passive Voice**: The subject **receives** the action. The focus is on the action, not the doer. (Structure: **Object + Helping Verb + V3 + by + Subject**)
    * **Example**: `The suspect was arrested by the police.`

**More Examples of Conversion**:
* **Active**: The lawyer will file the appeal.
    **Passive**: The appeal will be filed by the lawyer.
* **Active**: The committee is reviewing the policy.
    **Passive**: The policy is being reviewed by the committee.
* **Active**: Someone has stolen the evidence.
    **Passive**: The evidence has been stolen. (The doer is unknown or unimportant)

## 5. Modals: Expressing Attitude

Modals are auxiliary (helping) verbs that express ability, possibility, permission, or obligation.

* **Ability**: `can`, `could`
    * Examples: He **can** cite many precedents. He **could** argue brilliantly in his youth.
* **Permission**: `may`, `can`, `could`
    * Examples: **May** I approach the bench? **Can** I see that document?
* **Possibility**: `may`, `might`, `could`
    * Examples: The witness **might** be lying. The judgment **could** go either way.
* **Advice**: `should`, `ought to`
    * Examples: You **should** read the contract carefully. You **ought to** respect the court.
* **Obligation/Necessity**: `must`, `have to`
    * Examples: You **must** file the reply within 30 days. He **has to** appear in court tomorrow.

## 6. Prepositions: Showing Relationships

Prepositions link nouns/pronouns to other words in a sentence, showing relationships of time, place, and direction.

* **Prepositions of Time**:
    * **on**: Used for days and dates. (e.g., `on Monday`, `on October 2nd`)
    * **in**: Used for months, years, seasons, parts of the day. (e.g., `in December`, `in 2024`, `in the morning`)
    * **at**: Used for specific times. (e.g., `at 3 PM`, `at noon`)
* **Prepositions of Place**:
    * **in**: Inside an enclosed space. (e.g., `in the office`, `in the book`)
    * **at**: At a specific point or location. (e.g., `at the courthouse`, `at the bus stop`)
    * **on**: On a surface. (e.g., `on the table`, `on the second floor`)
* **Prepositions of Direction**:
    * **to**: Indicates movement toward a destination. (e.g., `He went to the court.`)
    * **into**: Indicates movement inside something. (e.g., `The lawyer walked into the room.`)
    * **from**: Indicates the point of origin. (e.g., `She came from her chamber.`)

## 7. Infinitives and Gerunds: Verbs in Disguise

### Infinitives (`to` + Verb)
The base form of a verb, often preceded by `to`. It can act as a noun, adjective, or adverb.

* **As a Noun (Subject/Object)**: `To argue` is his passion. He wants `to win`.
* **As an Adjective (modifies a noun)**: This is the best argument `to present`.
* **As an Adverb (modifies a verb/adjective)**: He stood up `to object`. The case is easy `to understand`.
* **Bare Infinitive** (without `to`): Used after modals and verbs like `let`, `make`, `see`, `hear`.
    * Examples: You must `obey` the law. She made him `confess`.

### Gerunds (Verb + `-ing`)
The `-ing` form of a verb that functions as a **noun**.

* **As a Subject**: `Filing` lawsuits can be expensive.
* **As an Object**: He enjoys `cross-examining` witnesses.
* **As an Object of a Preposition**: She is an expert in `drafting` contracts.

***

## 8. Articles: A/An, The

* **Indefinite Articles (`a`, `an`)**: Used for general, non-specific nouns.
    * `a` is used before a consonant sound. (e.g., `a lawyer`, `a university` [starts with 'y' sound])
    * `an` is used before a vowel sound. (e.g., `an FIR`, `an hour` [h is silent])
    * **Example**: I need to consult **a** lawyer. (any lawyer, not a specific one)

* **Definite Article (`the`)**: Used for specific, previously mentioned, or unique nouns.
    * **Example**: I consulted **the** lawyer we met yesterday. (a specific lawyer)
    * **Example**: **The** Chief Justice of India delivered the speech. (unique person)

* **Zero Article** (No article): Used with plural nouns in a general sense, proper nouns, and abstract nouns.
    * **Example**: `Lawyers` must be diligent. (general) `Justice` must prevail. (abstract)

## 9. Degrees of Comparison (Adjectives)

Adjectives change form to show comparison.

| Degree | Use | One Syllable | Two+ Syllables | Irregular |
| :--- | :--- | :--- | :--- | :--- |
| **Positive** | Describes one thing | `strong` | `important` | `good`, `bad` |
| **Comparative**| Compares two things | `stronger` | `more important`| `better`, `worse` |
| **Superlative**| Compares 3+ things | `strongest` | `most important`| `best`, `worst` |

**Examples**:
* This is a **strong** argument.
* His argument is **stronger than** the opposition's.
* This is the **strongest** point in the entire case.
* This evidence is **good**, but the other is **better**.

## 10. Editing and Omission


### Editing
You are given a sentence with one error. You must identify and correct it.

1.  **Incorrect**: The jury **are** unanimous in its decision.
    **Correct**: The jury **is** unanimous in its decision. (Jury as a single unit)
2.  **Incorrect**: He is senior **than** me in the profession.
    **Correct**: He is senior **to** me in the profession. (Certain words take 'to', not 'than')
3.  **Incorrect**: Neither the client nor his lawyers **was** present.
    **Correct**: Neither the client nor his lawyers **were** present. (Verb agrees with the nearest subject)

### Omission
You are given a sentence with a missing word. You must insert the correct word.

1.  **Sentence**: The judge did not agree ^ the lawyer's point.
    **Missing word**: `with`
    **Correct**: The judge did not agree **with** the lawyer's point.
2.  **Sentence**: He is accused ^ theft.
    **Missing word**: `of`
    **Correct**: He is accused **of** theft.
3.  **Sentence**: The meeting will be held ^ 4 PM and 6 PM.
    **Missing word**: `between`
    **Correct**: The meeting will be held **between** 4 PM and 6 PM.

# #Tricky Practice Q/A:
***

### 1. Subject, Predicate & Subject-Verb Agreement

**Q1. Find the error:** One of the most important legal documents **are** missing from the file.
* **Answer**: The error is **`are`**. It should be **`is`**.
* **💡 The Trick**: The subject of the sentence is "**One**," which is singular. The phrase "of the most important legal documents" is a prepositional phrase describing "One." The verb must agree with the singular subject, `One`, not the plural noun `documents`.

**Q2. Find the error:** Neither the witnesses nor the defendant **are** telling the whole truth.
* **Answer**: The error is **`are`**. It should be **`is`**.
* **💡 The Trick**: With `neither...nor` (or `either...or`), the verb agrees with the subject that is **closest** to it. Here, the closest subject is "the defendant," which is singular.

***

### 2. Noun Cases & Pronouns

**Q1. Choose the correct option:** The judge gave the file to my partner and (**I / me**).
* **Answer**: **`me`**
* **💡 The Trick**: To figure this out, remove the other person. You wouldn't say, "The judge gave the file to I." The pronoun is the object of the preposition "to," so it must be in the objective case (`me`).

**Q2. Choose the correct option:** **Who / Whom** did you give the document to?
* **Answer**: **`Whom`**
* **💡 The Trick**: A simple way to check is to rephrase the sentence as an answer: "I gave the document to **him**." Since you would use "him" (objective case), the correct question word is the objective `Whom`. If the answer were "he," you would use `Who`.

***

### 3. Tense

**Q1. Correct the sentence:** By the time the police arrived, the thief **left**.
* **Answer**: By the time the police arrived, the thief **had left**.
* **💡 The Trick**: When you have two actions that happened in the past, the action that happened **first** should be in the Past Perfect tense (`had + V3`). The thief leaving happened before the police arrived.

**Q2. Correct the sentence:** I **am living** in this city for five years.
* **Answer**: I **have been living** in this city for five years.
* **💡 The Trick**: When an action started in the past and is **still continuing** in the present, you must use the Present Perfect Continuous tense. "Am living" just describes the present moment.

***

### 4. Active & Passive Voice

**Q1. Change to Passive Voice:** Someone should have warned him.
* **Answer**: He **should have been warned**.
* **💡 The Trick**: The challenge here is the modal verb (`should have`). The passive structure for modals is **Modal + have + been + V3**. The "by someone" is usually omitted because it's vague.

**Q2. Change to Active Voice:** The case is being heard by the new judge.
* **Answer**: The new judge **is hearing** the case.
* **💡 The Trick**: "Is being heard" is the passive form of the Present Continuous tense. To make it active, you must convert it back to the active Present Continuous form (`is/am/are + V-ing`).

***

### 5. Infinitives & Gerunds

**Q1. Choose the correct option:** I stopped (**to smoke / smoking**) last year.
* **Answer**: **`smoking`**
* **💡 The Trick**: This tests the change in meaning.
    * `stop smoking` means to quit the habit. (Gerund)
    * `stop to smoke` means to stop what you are doing *in order to* have a cigarette. (Infinitive of purpose)

**Q2. Correct the sentence:** The lawyer made the witness **to answer** the question.
* **Answer**: The lawyer made the witness **answer** the question.
* **💡 The Trick**: After certain verbs like `make`, `let`, and `hear`, we use the **bare infinitive** (the verb without `to`).

***

### 6. Articles

**Q1. Fill in the blank:** He is studying at **___** university in London. (a / an)
* **Answer**: **`a`**
* **💡 The Trick**: The choice between `a` and `an` depends on the **sound**, not the letter. "University" starts with a 'y' sound (yoo-ni-ver-sity), which is a consonant sound.

**Q2. Fill in the blank:** **___** justice should be served. (A / The / No article)
* **Answer**: **No article**
* **💡 The Trick**: Abstract nouns (like justice, honesty, truth) used in a general sense do not take an article.

***

### 7. Degrees of Comparison

**Q1. Correct the sentence:** She is **wiser** than honest.
* **Answer**: She is **more wise** than honest.
* **💡 The Trick**: When you are comparing two different qualities **of the same person or thing**, you use `more` + the positive form of the adjective, even for one-syllable adjectives. We are not comparing her to someone else; we are comparing her wisdom to her honesty.

**Q2. Correct the sentence:** This is the **most perfect** solution.
* **Answer**: This is the **perfect** solution.
* **💡 The Trick**: Some adjectives like `perfect`, `unique`, `complete`, and `absolute` represent an ultimate state. They cannot be compared and do not have comparative or superlative forms.

***

### 8. Editing & Omission (Mixed Bag)

**Q1. Find the error:** The team of lawyers, along with their senior counsel, **are** preparing for the final hearing.
* **Answer**: The error is **`are`**. It should be **`is`**.
* **💡 The Trick**: The subject is "The team." Phrases like "along with," "as well as," or "in addition to" do not make the subject plural. The verb must agree with the main subject.

**Q2. Find the missing word:** He is addicted **^** gambling.
* **Answer**: The missing word is **`to`**.
* **💡 The Trick**: This tests your knowledge of **fixed prepositions**. The word "addicted" is always followed by "to."

## 📝 Essay and Story Writing

For a 10-mark question, you should demonstrate a clear understanding of the structure and key elements of the chosen format (essay or story).

### Essay Writing
An **essay** is a structured piece of writing that presents and develops an argument or idea on a specific topic. Its strength lies in its logical flow, coherence, and supporting evidence.

#### Structure of a Standard Essay
A good essay typically has three main parts:

1.  **Introduction (1 paragraph)**:
    * **Hook**: The first sentence that grabs the reader's attention (e.g., a surprising fact, a relevant quote, or a question).
    * **Background**: Briefly introduce the topic and its context.
    * **Thesis Statement**: The most important sentence of your essay. It clearly states your main argument or the central point you will prove in the essay.

2.  **Body (2-3 paragraphs)**:
    * Each paragraph focuses on a **single idea** that supports your thesis statement.
    * **Topic Sentence**: The first sentence of the paragraph, which states the main idea of that paragraph.
    * **Supporting Sentences**: Provide evidence, examples, facts, or explanations to develop the topic sentence.
    * **Concluding Sentence**: Summarizes the point of the paragraph or links it to the next one.

3.  **Conclusion (1 paragraph)**:
    * **Restate Thesis**: Rephrase your thesis statement in a new way.
    * **Summarize Main Points**: Briefly recap the main ideas from your body paragraphs.
    * **Final Thought**: End with a concluding remark, a suggestion, a prediction, or a powerful statement that leaves a lasting impression on the reader.

#### Example: Essay Topic - "Is Social Media More Harmful Than Beneficial?"

**Introduction Example**:
> In the 21st century, it is nearly impossible to imagine life without the constant hum of notifications from platforms like Instagram, Twitter, and Facebook. While these tools have undoubtedly connected the world in unprecedented ways, their detrimental effects on mental health and social discourse are becoming increasingly apparent. **Therefore, despite its benefits in communication, social media is ultimately more harmful than beneficial due to its addictive nature, its role in spreading misinformation, and its negative impact on self-esteem.**

*(This introduction has a hook, background, and a clear thesis statement in bold.)*

### Story Writing
A **story** is a narrative account of real or imaginary events. It focuses on entertaining the reader through elements like plot, character, and setting.

#### Key Elements of a Story
To write a compelling story, you must include these five core elements:

1.  **Plot**: The sequence of events in the story. It usually follows a structure:
    * **Exposition**: Introduction of characters and setting.
    * **Rising Action**: Events building up to the main conflict.
    * **Climax**: The most exciting point of the story; the turning point.
    * **Falling Action**: Events after the climax, leading to the resolution.
    * **Resolution**: The conclusion or end of the story.
2.  **Characters**: The people (or beings) who participate in the story's action. The **protagonist** is the main character, while the **antagonist** is the character or force in opposition.
3.  **Setting**: The time and place where the story occurs. Good setting descriptions help create the mood or atmosphere.
4.  **Conflict**: The central problem or struggle that the characters face. It can be internal (character vs. self) or external (character vs. character, nature, or society).
5.  **Theme**: The underlying message or main idea of the story (e.g., "courage in the face of adversity" or "the dangers of greed").

#### Example: Short Story Snippet
**Prompt**: Write a short story about a detective who finds an unusual clue.

> Detective Sharma stared at the crime scene, a pristine library owned by a reclusive millionaire. Everything was in its place—except for one thing. Tucked inside a dusty copy of "Moby Dick" was not a bookmark, but a single, perfectly preserved raven's feather. It was out of place, a dark slash of ink against the yellowed page. While his colleagues saw it as irrelevant, Sharma felt a chill. He knew the victim's last name was Poe. This wasn't just a clue; it was a message. The killer wasn't just a thief; they were a poet, and this was the first line of their dark verse. The real story, he realized, wasn't about what was stolen from the room, but what had been left behind.

*(This snippet establishes a character (Sharma), setting (library), conflict (murder mystery), and introduces the plot with a key clue.)*

## 📖 Paragraph and Précis Writing

This section tests your ability to write concisely and to summarize information accurately.

### Paragraph Writing
A **paragraph** is a self-contained unit of writing that deals with a single point or idea. It is a fundamental building block for any longer piece of writing.

#### Structure of a Good Paragraph
A well-structured paragraph follows the **T.E.E.** model:

* **T - Topic Sentence**: The first sentence that clearly states the main idea of the paragraph. Everything else in the paragraph must relate back to this sentence.
* **E - Explanation & Evidence**: The sentences that follow the topic sentence. They explain the main idea further and provide supporting details, examples, or facts to make it convincing.
* **E - Ending Sentence**: The last sentence that summarizes the idea or provides a transition to the next paragraph.

#### Example: Paragraph Topic - "The Importance of Time Management for Students"

> **(T)** Effective time management is arguably the most critical skill a student can develop for academic success. **(E)** Without a clear plan for allocating time to studies, assignments, and personal activities, students often find themselves overwhelmed and submitting subpar work. For instance, a student who schedules specific hours for revision is less likely to cram the night before an exam, leading to better retention and lower stress levels. This disciplined approach also prevents burnout by ensuring there is time for rest and recreation. **(E)** Ultimately, learning to manage time is not just about creating a schedule; it's about building a foundation for a balanced and successful academic career.

---

### Précis Writing
A **précis** (pronounced "pray-see") is a concise summary of a passage, written entirely in your **own words**. It should capture the essential ideas of the original text accurately and briefly, typically reducing it to about **one-third** of its original length.

#### Steps for Writing a Précis
1.  **Comprehend**: Read the original passage carefully two or three times to fully understand its meaning, main argument, and tone.
2.  **Identify**: Underline or note down the most important points and keywords. Distinguish the main ideas from examples or minor details.
3.  **Draft**: Write a first draft in your own words. Do not copy sentences from the original passage. Focus on conveying the author's core message.
4.  **Refine**: Check your draft against the original to ensure you haven't missed any key points or misinterpreted the meaning. Ensure it is a single, well-connected paragraph.
5.  **Edit**: Cut out any unnecessary words or redundant phrases. Check for grammar and spelling errors. Make sure it adheres to the required length (usually one-third).
6.  **Title**: Give your précis a suitable title that reflects the main subject of the passage.

#### Example: Précis Writing

**Original Passage (120 words)**:
> The advent of the internet has fundamentally transformed the landscape of modern communication. In the past, information flowed through centralized channels like newspapers and television broadcasts, which were controlled by a select few. This one-way model of communication offered little room for public discourse. Today, however, social media platforms and blogs have democratized the creation and dissemination of information. Any individual with an internet connection can share their views with a global audience, fostering a multi-directional flow of dialogue. While this has led to a more engaged citizenry, it has also created significant challenges, such as the rapid spread of misinformation and the erosion of trust in traditional institutions. The challenge for society is to harness the benefits of this new era while mitigating its risks.

**Précis Example**:

**Title**: The Internet's Impact on Communication

**Précis (41 words)**:
> The internet has shifted communication from a centralized, one-way model to a democratized system where anyone can share information globally. While this fosters greater public engagement, it also presents serious challenges like the spread of misinformation, requiring society to manage its risks.

## ✍️ Abstract Writing

An **abstract** is a brief, self-contained summary of a longer work, such as a research paper, thesis, or legal report. Its purpose is to give the reader a quick overview of the full document's content, helping them decide if it's relevant to their needs without having to read the entire piece.

#### Key Characteristics of a Good Abstract
* **Concise**: It is typically short, usually between 150-250 words.
* **Accurate**: It correctly reflects the purpose and content of the full paper.
* **Self-Contained**: It is understandable on its own, without reference to the main document.
* **Objective**: It reports the key findings without adding personal opinions or interpretations.

#### Structure of an Abstract
A well-written abstract generally follows a clear, four-part structure that answers four key questions:

1.  **Purpose/Motivation (Why did you do it?)**: Start by stating the principal objective or the problem being addressed. What is the research question?
2.  **Methodology (How did you do it?)**: Briefly describe the methods or approach used to conduct the research. Was it a survey, an analysis of case law, an experiment?
3.  **Results/Findings (What did you find?)**: Summarize the main results or key findings. This is often the most important part of the abstract.
4.  **Conclusion/Implications (What does it mean?)**: Conclude with a brief statement about the implications or significance of your findings. What is the main takeaway?

---
#### Example: Abstract Writing

**Hypothetical Research Paper Title**: The Impact of Remote E-Hearings on the Efficiency of District Courts in Delhi.

**Abstract**:
> **(Purpose)** This study investigates the impact of mandatory remote e-hearings on the operational efficiency of district courts in Delhi following the 2020 pandemic. The primary objective was to determine whether the transition to a virtual format improved or hindered the speed of case disposals. **(Methodology)** A quantitative analysis of case disposal rates was conducted using court data from 2018 to 2022. This was supplemented by qualitative data from structured interviews with 50 advocates to understand their procedural experiences. **(Results)** The findings indicate that while e-hearings initially caused a 30% decline in efficiency due to technical challenges, they led to a 15% increase in the disposal of interlocutory applications by 2022. However, the disposal rate for evidence-based trial matters remained significantly lower compared to pre-pandemic physical hearings. **(Conclusion)** The study concludes that e-hearings are effective for procedural matters but are less efficient for substantive trials. It recommends a hybrid model to optimize court efficiency.

## 🗒️ Note Writing (Note Making)

**Note writing**, or note making, is the process of recording key information from a source (like a lecture, textbook, or article) in a systematic and condensed format. The primary goal is to capture the core ideas, making it easier to understand, remember, and revise the material later.

#### Key Characteristics of Good Notes
* **Brevity**: Use phrases, keywords, and abbreviations instead of full sentences.
* **Organization**: Information should be structured logically using headings, subheadings, and indentation.
* **Clarity**: The notes should be easy to read and understand.
* **Relevance**: Only the most important points and supporting details should be included.

#### Structure and Method (Outline Method)
A common and effective method for note-making is the outline method, which uses a hierarchical structure.

1.  **Title**: Give a clear title that reflects the topic of the notes.
2.  **Main Headings**: Use numbers (1, 2, 3...) for the main topics.
3.  **Subheadings**: Use letters (a, b, c...) and indentation for points related to the main topics.
4.  **Further Details**: Use smaller roman numerals (i, ii, iii...) or further indentation for supporting details.
5.  **Use of Abbreviations**: Consistently use standard abbreviations to save time and space.
6.  **Key**: At the end, provide a key explaining the abbreviations and symbols used.

---
#### Example: Note Writing

**Source Passage**:
> The Basic Structure Doctrine is a legal principle evolved by the Supreme Court of India in the landmark case of *Kesavananda Bharati v. State of Kerala* (1973). This doctrine posits that while the Parliament of India has the power to amend the Constitution under Article 368, this power is not absolute. The Parliament cannot amend, abrogate, or destroy the "basic structure" or fundamental framework of the Constitution. The Supreme Court, however, has not provided an exhaustive list of what constitutes the basic structure. Over time, through various judgments, elements like the supremacy of the Constitution, the rule of law, judicial review, secularism, federalism, and the separation of powers between the legislature, executive, and judiciary have been recognized as part of it. The doctrine serves as a crucial check on the legislative power of the Parliament, ensuring the core ideals of the Constitution remain intact.
